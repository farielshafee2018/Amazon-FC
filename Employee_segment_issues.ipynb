```python
df4["sentiment_polarity"]= df3["cons_sentiment_polarity"]
```

```python
df4["sentiment_label"]=df3["cons_sentiment_label"]
```

```python
df4["Theme"]=df3["Theme"]
```

```python
import numpy as np
```

```python
df4["segment"] = np.select(
    [
        (df4["employee_status"] == "REGULAR") & (df4["employee_job_title"].str.contains("picker", case=False, na=False)),
        (df4["employee_status"] == "REGULAR") & (df4["employee_job_title"].str.contains("packer", case=False, na=False)),
        (df4["employee_status"] == "CONTRACT") & (df4["employee_job_title"].str.contains("picker", case=False, na=False)),
        (df4["employee_status"] == "CONTRACT") & (df4["employee_job_title"].str.contains("packer", case=False, na=False)),
        (df4["employee_status"] == "CONTRACT") & (df4["employee_job_title"].str.contains("associate|stower", case=False, na=False)),
        (df4["employee_status"] == "REGULAR") & (df4["employee_job_title"].str.contains("associate|stower", case=False, na=False)),
        df4["employee_job_title"].str.contains("management|managerghhj", case=False, na=False),
        df4["employee_job_title"].str.contains("IT|it|software", case=False, na=False),
        df4["employee_job_title"].str.contains("intern", case=False, na=False),
        df4["employee_job_title"].str.contains("engineer|engineering|senior|leader", case = False, na=False)
    ],
    [
        "regular picker",
        "regular packer",
        "contract picker",
        "contract packer",
        "other contract associates",
        "other regular associates",
        "management",
        "IT",
        "intern",
        "other senior"
    ],
    default="other"
)
```
