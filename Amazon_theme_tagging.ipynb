```python
import pandas as pd
```

This notebook finds themes in comments and tags the comments with themes

```python
df = pd.read_csv("/content/combined_sentiment_data.csv")
```

```python
df.head(2)
```

```python
theme_keywords = {
    'Insecurity': ['promotion', 'advance', 'insurance', 'retension','rehire','get rid','job security'],
    'Work type': ['hard','long hours','monotonous','training','shift','pain','injury','repetitive'],
    'Management Issues': ['untalented','management', 'treat','relation','attitude','politics','partiality','nepotism'],
    'Benefits':["low pay","compensation",'benefits','contract','bonus'],
    'Treatment':['inhuman','robot','break','personal','treatment']
}
```

```python
def tag_themes(text, theme_keywords):
    matched_themes = []
    for theme, keywords in theme_keywords.items():
        if any(keyword in str(text).lower() for keyword in keywords):
            matched_themes.append(theme)
    return ', '.join(matched_themes) if matched_themes else ''
```

```python
df['Theme'] = df['review_cons_cleaned'].apply(lambda x: tag_themes(x, theme_keywords))
```

```python
del df['transcript_cleaned']
del df['sentiment_polarity']
del df['sentiment_label']
```
